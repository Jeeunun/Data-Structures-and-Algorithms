### 격자판 최대합
5*5 격자판에 아래와 같이 숫자가 적혀있습니다. <br>
 <br>
N*N의 격자판이 주어지면 각 행의 합, 각 열의 합, 두 대각선의 합 중 가장 큰 합을 출력합니다. <br>

#### 입력설명
첫 줄에 자연수 N이 주어진다. (1<=N<=50)
두번재 줄부터 N줄에 걸쳐 각 줄에 N개의 자연수가 주어진다. 각 자연수는 100을 넘지 않는다.

#### 출력설명
최대합을 출력합니다.

10 13 10 12 15
12 30 30 28 11
11 25 30 53 15
19 27 29 37 27
19 13 30 13 10
